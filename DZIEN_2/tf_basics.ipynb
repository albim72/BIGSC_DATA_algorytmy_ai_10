```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train, x_test = x_train/255.0, x_test/255.0
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train[:10]
```

```python
#budowa modelu sieci neuronowej
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)), #warstwa redukcji wymiarów - przekształcenie macierzy w wektor
    tf.keras.layers.Dense(128,activation='relu'), #wartstwa ukryta z funkcją aktywacji relu
    tf.keras.layers.Dropout(0.2), #wartwa odrzucenia - 20% zmian w propagacji wstecznej
    tf.keras.layers.Dense(10) #warstwa wyjścia - 10 przypadków - 10 obrazów
])
```

```python
model.compile(optimizer="adam",
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy']
)

#propagacja wsteczna - algorytm zmiany wag: adam, funckja straty - entropia krzyzowa, metryki - ocen amodelu
```

```python
model.fit(x_train,y_train,epochs=10) #trening
```

```python
#ocena modelu
test_loss, test_acc = model.evaluate(x_test,y_test)
print(f"strata: {test_loss}") #błędy
print(f"accuracy - dokładnośc: {test_acc}") #prawdopodobieństwo prawidłowego rozwiązania
```
